```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import sympy
```

```python
## Example 1
# define function f(x)
def f(x):
    return 5*(x-1)**2

# define x and y
x = np.linspace(-20, 20, 100) # 100 equally spaced points on interval [-20,20]
y = f(x) # call function f(x) and set y to the function's return value

# Plot the function y=f(x)
plt.figure(figsize=(10, 10))
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('$y=5(x-1)^2$')
plt.show()
```

```python
## Question 1.1(concave function)
# Define function f(x)
f = lambda x,y: -(x**2 + y**2)

# Initialize figure
fig = plt.figure(figsize=(10, 10))
ax = fig.gca(projection='3d')

# Evaluate function
X = np.arange(-2, 2, 0.1)
Y = np.arange(-2, 2, 0.1)
X, Y = np.meshgrid(X, Y)
Z = f(X,Y)

# Plot the surface
surf = ax.plot_surface(X, Y, Z, cmap=cm.jet_r,
                       linewidth=0, antialiased=False) # colour map
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('$f(x,y) = (1-x)^2 + 100(y-x^2)^2$')
ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)
ax.set_zlim(-8, 0)
plt.show()

## Question 1.1(non-convex function in 3D)
# Define function f(x)
f = lambda x,y: -x**2 + y**2

# Initialize figure
fig = plt.figure(figsize=(10, 10))
ax = fig.gca(projection='3d')

# Evaluate function
X = np.arange(-2, 2, 0.1)
Y = np.arange(-2, 2, 0.1)
X, Y = np.meshgrid(X, Y)
Z = f(X,Y)

# Plot the surface
surf = ax.plot_surface(X, Y, Z, cmap=cm.jet_r,
                       linewidth=0, antialiased=False)
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('$f(x,y) = (1-x)^2 + 100(y-x^2)^2$')
ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)
ax.set_zlim(-5, 5)
plt.show()
```

```python
### Question 1.3(concave function)
## First derivative
# Define function z
x = sympy.Symbol('x')
y = sympy.Symbol('y')
z = -(x**2 + y**2)
zdx = z.diff(x)
zdy = z.diff(y)
#print ("dz/dx =",zdx,' ',"dz/dy =",zdy)

# define x and y
x_value = []
y_value = []
zdx_value = []
zdy_value = []

for i in np.arange(-4,4,0.1): # mesh
    x_value.append(i)
    y_value.append(i)
    zdx_value.append(zdx.subs('x',i))
    zdy_value.append(zdy.subs('y',i))

# Plot the first order derivative of the function
fig, (ax1, ax2) = plt.subplots(2, 1)
fig.suptitle('The first order derivative of the function $z = -(x^2+y^2)$', fontsize=12)
ax1.plot(x_value, zdx_value)
ax1.set(xlabel="X axis", ylabel="Z axis")
ax1.set_title(r'$\dfrac{\partial z}{\partial x}$')
ax2.plot(y_value, zdy_value)
ax2.set(xlabel="Y axis", ylabel="Z axis")
ax2.set_title(r'$\dfrac{\partial z}{\partial y}$')

fig.tight_layout()
fig.subplots_adjust(top=0.8) # arrange the space

plt.show()

## Second derivative
# Define function z
x = sympy.Symbol('x')
y = sympy.Symbol('y')
z = -(x**2 + y**2)
zdxx = zdx.diff(x)
zdxy = zdx.diff(y)
zdyy = zdy.diff(y)
zdyx = zdy.diff(x)

# define x and y
x_value = []
y_value = []
zdxx_value = []
zdxy_value = []
zdyx_value = []
zdyy_value = []

for i in np.arange(-4,4,0.1):
    x_value.append(i)
    y_value.append(i)
    zdxx_value.append(zdxx.subs('x',i))
    zdxy_value.append(zdxy.subs('x',i))
    zdyy_value.append(zdyy.subs('y',i))
    zdyx_value.append(zdyx.subs('y',i))

# Plot the second derivative of the function
fig, axarr = plt.subplots(2, 2)
fig.suptitle('The Second order derivative of the function $z = -(x^2+y^2)$', fontsize=12)
axarr[0, 0].plot(x_value, zdxx_value)
axarr[0, 0].set(xlabel="X axis", ylabel="Z axis")
axarr[0, 0].set_title(r'$\dfrac{\partial z}{\partial x^2}$')
axarr[0, 1].plot(x_value, zdxy_value)
axarr[0, 1].set(xlabel="X/Y axis", ylabel="Z axis")
axarr[0, 1].set_title(r'$\dfrac{\partial z}{\partial x \partial y}$')
axarr[1, 0].plot(x_value, zdyx_value)
axarr[1, 0].set(xlabel="X/Y axis", ylabel="Z axis")
axarr[1, 0].set_title(r'$\dfrac{\partial z}{\partial y \partial x}$')
axarr[1, 1].plot(x_value, zdxx_value)
axarr[1, 1].set(xlabel="Y axis", ylabel="Z axis")
axarr[1, 1].set_title(r'$\dfrac{\partial z}{\partial y^2}$')

fig.tight_layout()
fig.subplots_adjust(top=0.8)

plt.show()

### Question 1.3(non-convex function)
## First derivative
# Define function z
x = sympy.Symbol('x')
y = sympy.Symbol('y')
z = -x**2 + y**2
zdx = z.diff(x)
zdy = z.diff(y)
#print ("dz/dx =",zdx,' ',"dz/dy =",zdy)

# define x and y
x_value = []
y_value = []
zdx_value = []
zdy_value = []

for i in np.arange(-4,4,0.1):
    x_value.append(i)
    y_value.append(i)
    zdx_value.append(zdx.subs('x',i))
    zdy_value.append(zdy.subs('y',i))

# Plot the first derivative of the function
fig, (ax1, ax2) = plt.subplots(2, 1)
fig.suptitle('The first order derivative of the function $z = -x^2+y^2$', fontsize=12)
ax1.plot(x_value, zdx_value)
ax1.set(xlabel="X axis", ylabel="Z axis")
ax1.set_title(r'$\dfrac{\partial z}{\partial x}$')
ax2.plot(y_value, zdy_value)
ax2.set(xlabel="Y axis", ylabel="Z axis")
ax2.set_title(r'$\dfrac{\partial z}{\partial y}$')

fig.tight_layout()
fig.subplots_adjust(top=0.8)

plt.show()

## Second derivative
# Define function z
x = sympy.Symbol('x')
y = sympy.Symbol('y')
z = -x**2 + y**2
zdxx = zdx.diff(x)
zdxy = zdx.diff(y)
zdyy = zdy.diff(y)
zdyx = zdy.diff(x)

# define x and y
x_value = []
y_value = []
zdxx_value = []
zdxy_value = []
zdyx_value = []
zdyy_value = []

for i in np.arange(-4,4,0.1):
    x_value.append(i)
    y_value.append(i)
    zdxx_value.append(zdxx.subs('x',i))
    zdxy_value.append(zdxy.subs('x',i))
    zdyy_value.append(zdyy.subs('y',i))
    zdyx_value.append(zdyx.subs('y',i))

# Plot the second derivative of the function
fig, axarr = plt.subplots(2, 2)
fig.suptitle('The Second order derivative of the function $z = -x^2+y^2$', fontsize=12)
axarr[0, 0].plot(x_value, zdxx_value)
axarr[0, 0].set(xlabel="X axis", ylabel="Z axis")
axarr[0, 0].set_title(r'$\dfrac{\partial z}{\partial x^2}$')
axarr[0, 1].plot(x_value, zdxy_value)
axarr[0, 1].set(xlabel="X/Y axis", ylabel="Z axis")
axarr[0, 1].set_title(r'$\dfrac{\partial z}{\partial x \partial y}$')
axarr[1, 0].plot(x_value, zdyx_value)
axarr[1, 0].set(xlabel="X/Y axis", ylabel="Z axis")
axarr[1, 0].set_title(r'$\dfrac{\partial z}{\partial y \partial x}$')
axarr[1, 1].plot(x_value, zdxx_value)
axarr[1, 1].set(xlabel="Y axis", ylabel="Z axis")
axarr[1, 1].set_title(r'$\dfrac{\partial z}{\partial y^2}$')

fig.tight_layout()
fig.subplots_adjust(top=0.8)

plt.show()
```
